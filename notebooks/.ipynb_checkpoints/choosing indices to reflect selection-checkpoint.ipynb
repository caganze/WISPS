```python
df=wisps.COMBINED_PHOTO_SPECTRO_DATA.rename(columns={'grism_id': 'Names'})
```

```python
dflist=[]
for c in best:
    sl1=c.select(df, shapes_to_use=['T5-T9'])
    dflist.append(pd.concat([df[df.Names.isin(x)] for x in sl1]))
```

```python
import functools
t_dwarfs=functools.reduce(lambda x, y: pd.merge(x, y, how = 'inner', on='Names'), dflist)
```
