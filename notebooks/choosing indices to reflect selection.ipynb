```python
dflist=[]
for c in crts.values():
    sl1=c.select(df, shapes_to_use=['L0-L5', 'L5-T0', 'T0-T5', 'T5-T9'])
    dflist.append(pd.concat([df[df.Names.isin(x)] for x in sl1]))
```

```python
#print (dflist[0].Names.values)
import functools
t_dwarfs=functools.reduce(lambda x, y: pd.merge(x, y, how = 'inner', on='Names'), dflist)
```
