```python
import splat
import wisps
import pandas as pd

import glob
import numpy as np
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def format_column(col):
    return np.array(np.apply_along_axis(list, 0, col))[:,0]
```

```python
ids_cols=['H_2O-1/J-Cont',
 'H_2O-2/H_2O-1',
 'H-cont/H_2O-1',
 'CH_4/H_2O-1',
 'H_2O-2/J-Cont',
 'H-cont/J-Cont',
 'CH_4/J-Cont',
 'H-cont/H_2O-2',
 'CH_4/H_2O-2',
 'CH_4/H-Cont']
```

```python
dt=pd.read_pickle(wisps.OUTPUT_FILES+'/aegis_cut.pkl')
```

```python
indxmat=dt[ids_cols].apply(format_column, axis=1)
```

```python
x=np.array([i for i in indxmat])
```

```python
df=pd.DataFrame(x)
```

```python
df.columns=ids_cols
```

```python
df['Names']=dt['name'].values
df['spectra']=dt['spectra'].values
```

```python
def plot(s):
    try:
        sp=wisps.Source(name=s.filename)
        print (sp.flags, sp.spectral_type)
        if ((sp.flags[-3] ==1.0)& (sp.flags[-1]<0.1) & (sp.flags[-2]<0.1)): sp.plot()
    except KeyError: pass
```
